```python
"""
Python script to extract data from Magicbricks website. The following data will be extracted:
PropName,Address,PropType,PropDescription,Price,Area,NumberofBedrooms,NumberofBathrooms,Amenities,Latitude,Longitude
"""
```

```python
import requests
from bs4 import BeautifulSoup as bs

r=requests.get("https://www.magicbricks.com/property-for-rent/residential-real-estate?bedroom=1&proptype=Multistorey-Apartment,Builder-Floor-Apartment,Penthouse,Studio-Apartment,Service-Apartment&Locality=Katraj&cityName=Pune&BudgetMin=5000&BudgetMax=15000")
c=r.content

soup=bs(c,"html.parser")
all=soup.find_all("div",{"class":"m-srp-card SRCard"})

print(all[0].prettify())


```

```python
l=[]

for item in all:
    d={}
    proj=item.find("span",{"class":"hidden"})
    title=item.find("meta", itemprop="name")
    try:
        d["Name"]=proj["data-projectname"] if proj["data-projectname"] else title["content"]
    except:
        pass
    try:
        d["Address"]=item.find("meta", itemprop="addressLocality")["content"]
    except:
        d["Address"]=None
    d["Type"]=proj["data-propertyval"]
    try:
        d["Description"]=item.find("meta", itemprop="description")["content"]
    except:
        d["Description"]=None
    d["Price"]=proj["data-price"]
    try:
        d["Area"]=item.find("meta", itemprop="floorSize")["content"]
    except:
        d["Area"]=None
    d["No. of Bedrooms"]=proj["data-bedroom"]
    d["No. of Bathrooms"]=proj["data-bathroom"]
    try:
        d["Latitude"]=item.find("meta", itemprop="latitude")["content"]
    except:
        d["Latitude"]=None
    try:
        d["Longitude"]=item.find("meta", itemprop="longitude")["content"]
    except:
        d["Longitude"]=None
    
    l.append(d)
    
```

```python
import pandas
df=pandas.DataFrame(l)
```
